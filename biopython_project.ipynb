```python
from Bio import SeqIO
from collections import Counter
import matplotlib.pyplot as plt

# Ler as sequências dos arquivos FASTA
sequences_18s_human = open('/content/drive/MyDrive/UDEMY/biopython/18s_humano.fasta').read()

# Contar a quantidade pares de nucleotídeos nas sequências
count_18s = {}

nucleotide = ['A', 'T', 'C', 'G']

for i in nucleotide:
  for j in nucleotide:
    count_18s[i + j] = 0

sequences_18s_human = sequences_18s_human.replace('\n', '')

for k in range(len(sequences_18s_human) - 1):
  count_18s[sequences_18s_human[k] + sequences_18s_human[k + 1]] += 1

count_18s
```

```python
# Ler as sequências dos arquivos FASTA
sequences_16s_bacteria = open('/content/drive/MyDrive/UDEMY/biopython/16s_bacteria.fasta').read()

# Contar a quantidade pares de nucleotídeos nas sequências
count_16s = {}

nucleotide = ['A', 'T', 'C', 'G']

for i in nucleotide:
  for j in nucleotide:
    count_16s[i + j] = 0

sequences_16s_bacteria = sequences_16s_bacteria.replace('\n', '')

for k in range(len(sequences_16s_bacteria) - 1):
  count_16s[sequences_16s_bacteria[k] + sequences_16s_bacteria[k + 1]] += 1

count_16s
```

# Visualização dos Dados

## Pares de nucleotídeos do genes 18s do ser humano

```python
# Extrair as chaves e valores do dicionário
labels = list(count_18s.keys())
values = list(count_18s.values())

# Criar o gráfico de barras
fig, ax = plt.subplots()
ax.bar(labels, values)

# Configurar o título e os rótulos dos eixos
ax.set_title('Contagem de Pares de Nucleotídeos no gene 18s')
ax.set_xlabel('Pares de Nucleotídeos')
ax.set_ylabel('Contagem')

# Rotacionar os rótulos do eixo x para melhor legibilidade
plt.xticks(rotation=45, ha='right')

# Exibir o gráfico
plt.show()
```
